```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Load your dataset (adjust the path as necessary)
file_path = r'C:\Users\ievav\vehicles.csv'  # Adjust the file path as needed
vehicles_df = pd.read_csv(file_path)

# Calculate mean and standard deviation
mean = vehicles_df['Combined MPG'].mean()
std_dev = vehicles_df['Combined MPG'].std()

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(vehicles_df['Combined MPG'], bins=50, edgecolor='black', alpha=0.6, density=True)

xmin, xmax = plt.xlim()  # Get the current x-axis limits
x = np.linspace(xmin, xmax, 100)  # Generate 100 points between the limits

# Calculate the normal distribution for these x values
p = stats.norm.pdf(x, mean, std_dev)  # Probability density function
# Overlay the normal distribution curve
plt.plot(x, p, 'k', linewidth=2)
plt.title('Histogram de MPG combinado con Curva Normal')
plt.xlabel('Valor de MPG')
plt.ylabel('Densidad')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

¿Cuál o cuáles de las variables tienen una distribución casi normal? ¿Cómo lo sabes?

```python
# Sé que al superponer una curva de distribución normal sobre un histograma 
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
exponential_10 = np.random.exponential(scale=10, size=10)
exponential_100 = np.random.exponential(scale=10, size=100)
print(exponential_10)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # Create a subplot (1 row, 2 columns, 1st subplot)
plt.hist(exponential_10, bins=100, edgecolor='black', alpha=0.7)
plt.title('Histograma de 10 números distribuidos exponencialmente')
plt.xlabel('Valor')
plt.ylabel('Recuento')
plt.grid(axis='y', alpha=0.75)

# Histogram for the sequence of size 100
plt.subplot(1, 2, 2)  # Create a subplot (1 row, 2 columns, 2nd subplot)
plt.hist(exponential_100, bins=100, edgecolor='black')
plt.title('Histograma de 100 números distribuidos exponencialmente')
plt.xlabel('Valor')
plt.ylabel('Recuento')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
# Show the plot
plt.show()
```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia. 

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
mean = 10
lambda_value = 1/10
x = 15
probability = 1 - np.exp(-lambda_value * x)
```

¿Cuál es la probabilidad de que el cliente pase más de 15 minutos?

```python
print(probability)
```
